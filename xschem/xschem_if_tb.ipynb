```python
import os
import sys
import pathlib
import functools
import operator
foldr = lambda func, acc, xs: functools.reduce(lambda x, y: func(y, x), xs[::-1], acc)
```

```python
dir = './'
flist = []
for root, dirs, files in os.walk(dir):
    for f in files:
        if os.path.splitext(f)[1] == '.sch':
            fullpath = os.path.join(root, f)
            flist.append(fullpath)
```

```python
# Convert to netlist with xschem

fn = 'xcp_osc_tb.sch'
fn0 = foldr(operator.add,'',fn.split('.')[:-1])
print(fn0)
os.system(f'xschem -n -x -q {fn}')

fn1 = f'{fn0}.spice'
fp = open(fn1,'r')
ilines = fp.readlines()
fp.close()

# Run ngspice
os.system(f'ngspice {fn0}.spice')
```
